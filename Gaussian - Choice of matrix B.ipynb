For $a=2/3$,
the matrix $B=-\Delta$ should not be enough to take into account high order of regularity.
In particular, for $p=2$ the boundaries conditions changes.
Here we show that for a gaussian however it doesn't change anything.

## Fisrt Test

$$
D = B_1^{1/2}=(-\Delta)^{1/2}, \qquad p=2, \qquad D^2 = (-\Delta) \;.
$$

```python
exp0 = AbelSolver(a=1.5,p=2,methB='constant')
exp0.Slope(interval=(-3,-1),reg_param=(-13,-11),style='gauss')
```

## Second Test

$$
D = (B_2)^{1/4}, \qquad p=2, \qquad D^2 = B_2^{1/2} \;.
$$

```python
exp1 = AbelSolver(a=1.5,p=2,methB='explicit')
exp1.Slope(interval=(-3,-1),reg_param=(-14,-11),style='gauss')
```
