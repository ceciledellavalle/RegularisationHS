```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.linalg import inv,pinvh,eig,eigh
%matplotlib inline
```

```python
# Parameters
nx = 100 # signal size
dx = 1/nx
```

# Inverse Problem - Regularisation with derivative

$$a=1$$

```python
# Transformation matrix
a = 1
T = dx*np.tri(nx)
```

```python
# Derivative
tTT = T.transpose().dot(T)
B = 2*nx**2*np.diag(np.ones(nx)) - nx**2*np.diag(np.ones(nx-1),1) - nx**2*np.diag(np.ones(nx-1),-1)
B[0,0]= nx**2
# Test
Id = B.dot(tTT)
Id[Id<10**-10] =0
np.linalg.norm(Id-np.eye(nx))
```

```python
# Power of B
D   = B
tDD = B.transpose().dot(B)
```

```python
# Synthetic Data
t  = np.linspace(0,1,nx)
x  = np.exp(-(t-0.5)**2/0.1**2)
y  = T.dot(x)
no = np.random.randn(nx)
no = 0.1*no*np.linalg.norm(y)/np.linalg.norm(no)
yd = y + no
plt.plot(t,x)
plt.plot(t,y)
plt.plot(t,yd)
```

```python
q = 2 # regularity of x
p = (q+1)//2 + a # maximal order of derivative
```

```python
# inversion directe
alpha = 10**-5.5
xd    = np.linalg.inv(tTT + alpha*B).dot(np.transpose(T).dot(yd))
xd[xd<0] = 0
plt.plot(t,x)
plt.plot(t,xd)
```

### Conjugate gradient method

```python
q     = 2 # regularity of x
p     = (q+1)//2 + a # maximal order of derivative
alpha = 0.000000001
```

```python
# gradient conjugÃ©
# Fonction Quadratique 
def Conjugate(x0,f,df,ddf,eps=10**-5,Niter=500):
    # Init
    xk = x0.copy()
    nx = x0.size
    gk = df(xk)
    loss = []
    loss1 = []
    for k in range(Niter):
        if np.linalg.norm(gk) < eps :
            print("Number of iterations {}".format(k))
            print("Final error eps = {}".format(f(xk)))
            return xk, loss, loss1
        elif k==0 :
            wk = gk
        else :
            alphak = np.dot(gk,ddf.dot(wk))/np.dot(wk,ddf.dot(wk))
            wk = gk - alphak*wk
        rhok = np.dot(gk,wk)/np.dot(wk,ddf.dot(wk))
        xk -= rhok*wk
        gk = df(xk)
        loss.append(f(xk))
        loss1.append(np.linalg.norm(df(xk)))
    print("Final error eps = {}".format(f(xk)))
    print("Max iteration of {} reached".format(Niter))
    return xk, loss, loss1
```

```python
def f(x):
    fx = 0.5*np.linalg.norm(T.dot(x)-yd)+ 0.5*alpha*np.linalg.norm(D.dot(x))
    return fx

def df(x):
    dfx = tTT.dot(x) - np.transpose(T).dot(yd) + alpha*tDD.dot(x)
    return dfx
    
ddf = tTT + alpha*tDD
```

```python
x0 = np.zeros(nx)
```

```python
xd, loss, lossd = Conjugate(x0,f,df,ddf)
```
