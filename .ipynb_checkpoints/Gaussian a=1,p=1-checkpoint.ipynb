Here we choose a function with $q= +\infty$,
and therefore $p$ can be arbitrary large.

For a given $p$, then $x \in X_{r,q}$ with $q = 2p+a$.
If $a=1$ and $p=1$, then the slope is $q/(q+a) =0.75$.
If $a=1$ and $p=2$, then the slope is $q/(q+a) =0.85$.

### Off gaussian, a=1, p=1

```python
exp1 = AbelSolver(a=1,p=1)
```

```python
x,y,yd = exp1.DataGen(noise=10**-4,style='offgauss',export=True)
```

```python
xadp = exp1.Solver(x,yd,reg_inf=10**-50,reg_sup=10**-5,export=True)
```

```python
exp1.Slope(style='gauss')
```
