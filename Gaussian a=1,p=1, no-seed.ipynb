For a given $p$, then $x \in X_{r,q}$ with $q = 2p+a$.

If $a=1$ and $p=1$, then the slope is $q/(q+a) =0.75$.

If $a=1$ and $p=2$, then the slope is $q/(q+a) =0.85$.
